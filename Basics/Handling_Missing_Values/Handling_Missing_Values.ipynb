## Approach #4: Using Imputer

## strategy to handle missing values can be:
<ul> <li>mean</li>
     <li>median</li>
     <li>most_frequent</li>
</ul>

```python
from sklearn.impute import SimpleImputer

imputer_mean = SimpleImputer(missing_values = np.nan, strategy= 'mean', fill_value=None, verbose=0, copy=True)
imputer_median = SimpleImputer(missing_values = np.nan, strategy= 'median', fill_value=None, verbose=0, copy=True)
imputer_most_frequent = SimpleImputer(missing_values = np.nan, strategy= 'most_frequent', fill_value=None, verbose=0, copy=True)
```

```python
imputer_data_mean = imputer_mean.fit_transform(data[cols])
imputer_data_median = imputer_median.fit_transform(data[cols])
imputer_data_most_frequent = imputer_most_frequent.fit_transform(data[cols])
```

```python
imputer_data_mean
```

```python
imputer_data_mean = pd.DataFrame(data = imputer_data_mean, columns=cols)
imputer_data_median = pd.DataFrame(data = imputer_data_median, columns = cols)
imputer_data_most_frequent = pd.DataFrame(data = imputer_data_most_frequent, columns = cols)
```
